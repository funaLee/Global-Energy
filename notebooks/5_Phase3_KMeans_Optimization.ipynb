# Notebook 5: Phase 3 K-Means Optimization

**Goals:**
- Use K-Means to cluster countries based on economic/energy profile.
- Train specialized models for each cluster.
- Compare Global Model vs Cluster-specific Models.

```python
import pandas as pd
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
import sys
import os

sys.path.append(os.path.abspath(os.path.join('../src')))
from preprocessing import load_data

# Load basic data for clustering
df = load_data('../data/processed/common_preprocessed.csv')

# Select Clustering Features (e.g., mean values per country)
clustering_feats = ['gdp_per_capita', 'Access to electricity (% of population)']
df_country = df.groupby('Entity')[clustering_feats].mean()

# Cluster
kmeans = KMeans(n_clusters=3, random_state=42)
df_country['Cluster'] = kmeans.fit_predict(StandardScaler().fit_transform(df_country))

print("Cluster counts:")
print(df_country['Cluster'].value_counts())
```
